```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import precision_recall_curve, roc_curve, roc_auc_score
from sklearn.metrics import average_precision_score, precision_score, recall_score
```

```python
df=pd.read_csv('../02- Anomaly Detection/credit_card.csv')
df.head()
```

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2018,stratify=df['Class'])
```

```python
num_pip=Pipeline(steps=[
                     ('scaler',StandardScaler())

    
])

X_train_final=num_pip.fit_transform(X_train)
X_test_final=num_pip.transform(X_test)
```

```python
def Anomaly_score(X_original,X_reconstructed):
    loss=np.sum(((X_original-X_reconstructed)**2),axis=1)
    loss=(loss-np.min(loss))/(np.max(loss)-np.min(loss))
    
    return loss
```

```python
def evaluate_score(y_train,anomaly_score,return_df='False'):
    df= pd.DataFrame({'True lables': y_train,'Anomally Scores':anomaly_score}).reset_index()
    df=df.drop(columns='index',axis=1)
     ## Prec_recall_thre
    precision,recall,thresholds=precision_recall_curve(y_true=df['True lables'],y_score=df['Anomally Scores'])
    average_precision=average_precision_score(y_true=df['True lables'],y_score=df['Anomally Scores'])
    ## ploting
    plt.step(recall,precision,color='k')
    plt.fill_between(recall,precision,color='grey')
    plt.title(f'Average Precition Score = {average_precision:.2f}')
    plt.xlabel('Recall',fontsize=14,color='k')
    plt.ylabel('Precision',fontsize=14,color='k')
    plt.show()
    #ROC
    fpr,tpr,threshold=roc_curve(y_true=df['True lables'],y_score=df['Anomally Scores'])
    auc=roc_auc_score(y_true=df['True lables'],y_score=df['Anomally Scores'])
     ##Plotting
    plt.plot(fpr,tpr,color='k',linewidth=2)
    plt.plot([0,1],[0,1],color='k',linewidth=2,linestyle='--')
    plt.scatter(0,0,color='r',s=20)
    plt.scatter(1,1,color='r',s=20)
    plt.title(f'Roc Auc Score = {auc:.2f}')
    plt.xlabel('False Positive Rate',fontsize=14,color='k')
    plt.ylabel('True Positive RatePrecision',fontsize=14,color='k')
    plt.grid()
    plt.show()
    if return_df=='True':
        return df
```

```python
def plot_two_components(X_reduced,y_true,algo_name):
    df_two_comp=pd.DataFrame(np.c_[X_reduced[:,[0,1]],y_true],columns=['First Vector','Second Vector','Label'])
    ## plot
    sns.lmplot(x='First Vector',y='Second Vector',data=df_two_comp,hue='Label',fit_reg=False)
    plt.title(f'Using {algo_name} Algorithm',color='k')
    plt.show()
```

### PCA

```python
pca_30=PCA(n_components=30,random_state=45)
X_train_reduced_pca_30=pca_30.fit_transform(X_train_final)
X_train_reconstructed_pca_30=pca_30.inverse_transform(X_train_reduced_pca_30)
plot_two_components(X_train_reduced_pca_30,y_train,'PCA nd-30')
X_train_pca_30_score=Anomaly_score(X_train_final,X_train_reconstructed_pca_30)
evaluate_score(y_train,X_train_pca_30_score,return_df='True')
```

```python
pca_27=PCA(n_components=27,random_state=2018)
X_train_reduced_pca_27=pca_27.fit_transform(X_train_final)
X_train_reconstructed_pca_27=pca_27.inverse_transform(X_train_reduced_pca_27)
plot_two_components(X_train_reduced_pca_27,y_train,'PCA nd-27')
X_train_pca_27_score=Anomaly_score(X_train_final,X_train_reconstructed_pca_27)
df_pca_27=evaluate_score(y_train,X_train_pca_27_score,return_df='True')
df_pca_27
```
